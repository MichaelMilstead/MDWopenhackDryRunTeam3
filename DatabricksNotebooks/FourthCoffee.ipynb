```
# Create Transactions with correct date types
sqlContext.sql("select lower(TransactionID) as TransactionID, lower(CustomerID) as CustomerID, lower(MovieID) as MovieID, CAST(UNIX_TIMESTAMP(CAST(RentalDate AS String), 'yyyyMMdd') AS TIMESTAMP) as RentalDate, CAST(UNIX_TIMESTAMP(CAST(RentalDate AS String), 'yyyyMMdd') AS TIMESTAMP) as ReturnDate, RentalCost, LateFee, RewindFlag, CreatedDate, UpdatedDate from Transactions").repartition(1).write.format("csv").mode('overwrite').option("header","true").save("/mnt/datalake1/fourthcoffee-rentals/Transactions")
```
